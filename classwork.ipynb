# linear regression
 predicting marks based on study hours.

```python
# Simple Linear Regression without library

hours = [1, 2, 3, 4, 5]
marks = [20, 35, 50, 65, 80]

# calculate average slope
slope = (marks[-1] - marks[0]) / (hours[-1] - hours[0])
intercept = marks[0] - slope * hours[0]

# prediction
new_hour = 6
predicted_marks = slope * new_hour + intercept

print("Predicted marks for 6 hours study:", predicted_marks)
```

# logistic regression:
Predicting pass or fail based on marks.


```python
# Simple classification logic (Pass/Fail)

marks = int(input("Enter marks: "))

if marks >= 40:
    print("Result: PASS")
else:
    print("Result: FAIL")
```

# confusion matrix:
Comparing actual result vs predicted result.

```python
actual = [1, 0, 1, 1, 0]
predicted = [1, 0, 0, 1, 0]

tp = tn = fp = fn = 0

for i in range(len(actual)):
    if actual[i] == 1 and predicted[i] == 1:
        tp += 1
    elif actual[i] == 0 and predicted[i] == 0:
        tn += 1
    elif actual[i] == 0 and predicted[i] == 1:
        fp += 1
    else:
        fn += 1

print("True Positive:", tp)
print("True Negative:", tn)
print("False Positive:", fp)
print("False Negative:", fn)
```

# decision tree:
Decide whether to approve a loan.

```python
age = int(input("Enter age: "))
income = int(input("Enter income: "))

if age >= 25:
    if income >= 50000:
        print("Loan Approved")
    else:
        print("Loan Rejected due to low income")
else:
    print("Loan Rejected due to age")
```

# KNN:
Classify student as weak or strong based on neighbors

```python
marks = [30, 35, 40, 70, 75, 80]
labels = ["Weak", "Weak", "Weak", "Strong", "Strong", "Strong"]

new_mark = 45

# nearest value logic
nearest = marks[0]
index = 0

for i in range(len(marks)):
    if abs(marks[i] - new_mark) < abs(nearest - new_mark):
        nearest = marks[i]
        index = i

print("Student Category:", labels[index])
```
